```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.fft import rfft, rfftfreq, irfft
from tqdm import tqdm
```

```python
def lowband_pass(data, data_freq, cutoff_freq):
    n = len(data)
    yf = rfft(data)
    xf = rfftfreq(n, 1/data_freq)
    yf[xf > cutoff_freq] = 0
    return irfft(yf)
```

```python
x = np.linspace(0, 0.5, 1000)
y = np.sin(25*x*2*np.pi) + np.sin(1000*x*2*np.pi) * 0.2
```

```python
plt.plot(x, np.sin(x*2*np.pi))
plt.show()
```

```python
plt.plot(x, y)
plt.xlim(0, 0.5)
plt.show()
```

```python
n = len(y)
yf = rfft(y)
plt.plot(yf)
plt.show()
```

```python
xf = rfftfreq(n, 1/2000)
yf[xf > 100] = 0
plt.plot(yf)
plt.show()
```

```python
y_clean = irfft(yf)
plt.plot(y_clean)
plt.show()
```
